Applied K means Clustring without sckit-learn using pandas only

````python
import pandas as pd
import numpy as np
````

````python
df = pd.read_excel("P-AI_Data-Set.xlsx")
test = pd.read_excel("test.xlsx")
````

````python
test
````

````python
x = df['Salary'].tolist()
````

````python
x
````

````python
#del both salary columns
del(df['Salary'])
del(test['Salary'])
````

````python
df
````

````python
dummy1 = pd.get_dummies(df, drop_first=True)
        
````

````python
dummy1
        
````

````python
dummy2 = pd.get_dummies(test, drop_first=True)
````

````python
dummy2
````

````python
l1 = []

for i in dummy1.columns:
    if i not in dummy2.columns:
        l1.append(i)
    

for i in l1:
    dummy2[i] = 0  
````

````python
import math
import statistics
distance = []
salary2 = []
k = 3
for i in range(len(dummy2)):
    distt = []
    for j in range(len(dummy1)):
      #finding ecudilian distance 
        d = math.dist(dummy1.iloc[j],dummy2.iloc[i])
        distt.append(d)
    da = pd.DataFrame({"distance": distt,"salary": x})
    da.sort_values(by = 'distance' ,inplace = True)
    most = []
    for d in range(k):
        most.append(da.iloc[d]['salary'])
        
    if(statistics.mode(most)  >= 35000):  # which ever value (greater than 35000 in this case) appear most in k neighbours (eg: k = 3 ) it will give 1 to it otherwise 0
        salary2.append(1)
    else:
        salary2.append(0)
            
        
````

````python
test['Salary'] = salary2
````
